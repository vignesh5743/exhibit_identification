```python
import pandas as pd
import numpy as np
import random

# Define the number of records
num_records = 1000

# Initialize lists to store generated data
machine_id = []
floor = []
type_ = []
current = []
motion = []
vibration = []
classification = []

# Generate random data
for _ in range(num_records):
    # Generate machine_id (1 to 10)
    mid = random.randint(1, 10)
    machine_id.append(mid)
    
    # Generate floor (1 or 2)
    fl = random.randint(1, 2)
    floor.append(fl)
    
    # Generate type (1, 2, or 3)
    t = random.randint(1, 3)
    type_.append(t)
    
    # Generate current, allowing for negative values for testing
    if t in [1, 2]:
        curr = round(random.uniform(-2.0, 4.0), 2)  # Random current value between -2.0 and 4.0
    else:
        curr = 0  # No current for type 3
    current.append(curr)
    
    # Generate motion (0 or 1)
    mot = random.randint(0, 1)
    motion.append(mot)
    
    # Generate vibration as 0 or 1
    vib = random.randint(0, 1)
    vibration.append(vib)
    
    # Classification based on type
    if t in [1, 2]:
        classification.append(1)
    else:
        classification.append(0)

# Create a DataFrame
df = pd.DataFrame({
    'Machine_id': machine_id,
    'Floor': floor,
    'Type': type_,
    'Current': current,
    'Motion': motion,
    'Vibration': vibration,
    'Class': classification
})

# Save to CSV file
df.to_csv('generated_records5.csv', index=False)

print("Generated records saved to 'generated_records.csv'")

# Load the dataset
df = pd.read_csv(r'generated_records5.csv')

# Define a function to update the class based on the conditions
def update_class(row):
    if row['Current'] < 0:
        return 'Alert'
    elif row['Type'] == 1 and row['Current'] <= 0:
        return 'Alert'
    elif row['Type'] == 2 and row['Current'] > 0 and row['Vibration'] == 0:
        return 'Alert'
    elif row['Current'] == 0 and row['Type'] != 3:
        return 'Alert'
    elif row['Type'] == 3 and row['Vibration'] == 1:
        return 'Normal'
    elif row['Type'] == 3 and (row['Current'] == 0 or row['Vibration'] == 0):
        return 'Alert'
    else:
        return 'Normal'

# Apply the function to update the class column
df['Class'] = df.apply(update_class, axis=1)

# Save the updated dataset to a new CSV file
df.to_csv('updated_records7.csv', index=False)

print("Updated records saved to 'updated_records.csv'")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

df = pd.read_csv(r'updated_records7.csv')

label_encoder = LabelEncoder()
df['Class'] = label_encoder.fit_transform(df['Class'])

X = df[['Machine_id', 'Floor', 'Type', 'Current', 'Motion', 'Vibration']]
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_model = SVC(kernel='linear', random_state=42) 
svm_model.fit(X_train_scaled, y_train)

y_pred = svm_model.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
import joblib

# Save the trained model and scaler
joblib.dump(svm_model, 'svm_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
